```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn
from sklearn.cluster import KMeans
from scipy.cluster import hierarchy
%matplotlib inline
```

```python
df = pd.read_csv("Mall_Customers.csv")
# df.head()
# df.shape
# df.info()
# df.describe()
# df.describe(include=['O'])
# print(df.columns)
df["Genre"].replace(["Male", "Female"], [0,1], inplace=True)
```

```python
f1 = df["Annual Income (k$)"].values #X value
f2 = df["Spending Score (1-100)"].values #y value
X = np.vstack((f1, f2)).T
X.shape
```

```python
den = hierarchy.dendrogram(hierarchy.linkage(X, method='ward'))
```

```python
err= []
for i in range(1,10):
    km = KMeans(n_clusters=i, init='k-means++')
    km.fit(X)
    err.append(km.inertia_)
plt.plot(range(1,10), err)
```
